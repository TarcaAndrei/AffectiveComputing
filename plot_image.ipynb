```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Load the .npy file (replace with your file path)
file_path = 'img_arrs/img_arr_0.npy'  # Path to your .npy file containing the color image
image = np.load(file_path)

# Check the shape of the image (should be something like (height, width, 3) for RGB)
print(f"Image shape: {image.shape}")

new_width = 200
new_height = 150

# Resize the image using OpenCV
image = cv2.resize(image, (new_width, new_height))

# Plot the color image
plt.imshow(image)
plt.title("Color Image from .npy file")
plt.axis('off')  # Hide axis ticks
plt.show()

cropped_image = image[10:100, 30:90]
# Check the shape of the image (should be something like (height, width, 3) for RGB)
print(f"Image shape: {cropped_image.shape}")

# Plot the color image
plt.imshow(cropped_image)
plt.title("Color Image from .npy file")
plt.axis('off')  # Hide axis ticks
plt.show()

```
